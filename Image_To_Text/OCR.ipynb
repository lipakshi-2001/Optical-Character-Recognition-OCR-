# Optical Character Recognition (OCR)
### Image To Text Conversion 

## Import Modules

```python
import matplotlib.pyplot as plt
import PIL
import pytesseract
import re
%matplotlib inline
```

## Load the Image

### Noise Reduction

```python
# Import Dependencies
import cv2
from PIL import Image
```

```python
# Reading Image
img = cv2.imread('Intensity.jpeg')
print(img)
```

```python
# Apply Image Denoising
dst = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)
print(dst)
```

```python
# Plot Original and Denoised Image
titles = ["Original Image", "Denoised Image"]
images = [img,dst]
plt.figure(figsize=(20,20))
for i in range(2):
    plt.subplot(2,3,i+1) , plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]) , plt.yticks([])
```

```python
# Loading Image For Text Extraction Using PIL 

img_new = PIL.Image.open('Intensity.jpeg')
plt.imshow(img_new)
```

## Convert Image to Text

```python
# Config
pytesseract.pytesseract.tesseract_cmd = r'/usr/local/bin/tesseract'
```

```python
text_data = pytesseract.image_to_string(img_new.convert('RGB'), lang='eng')
```
